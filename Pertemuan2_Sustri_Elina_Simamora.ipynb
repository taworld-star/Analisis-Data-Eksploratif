```python
dFrame=pd.DataFrame({'F':np.linspace(1,10,10)})
dFrame=pd.concat([dFrame,pd.DataFrame(np.random.randn(10,5), columns=list('EDBCA'))],axis=1)
dFrame
```

```python
#Definisikan nilai dengan warna
def Warna(value):
  if value<0:
    color='teal'
  elif value>0:
    color='green'
  else:
    color='beige'
  return 'color:%s'%color
```

```python
S=dFrame.style.applymap(Warna,subset=['A','B','C','D','E'])
S
```

```python
#Higlight Bilangan Maximum dengan warna orange, minimum dengan warna hijau
def higlight(x):
  Maksimum=x==x.max()
  Minimum=x==x.min()
  return ['background-color:orange' if v else '' for v in Maksimum]

def highlight2(x):
  Minimum =x==x.min()
  return['background-color:green' if v else '' for v in Minimum]
```

Axis: Sumbu untuk grafik
Nilai Max dan Min di lihat per kolom

```python
styled_df=dFrame.style.apply(higlight,axis=0).apply(highlight2,axis=0)
#styled_df.applymap(lambda x:'color:red' if pd.isnull(x) else '') #highlight NaNs with red color
styled_df
```

**DISTRIBUSI**

scipy: digunakan untuk melakukan uji statistika, juga dapat digunakan untuk melakukan pemodelan statistika.
bin(pembagian)
kde (border line)


```python
import matplotlib.pyplot as plt
from IPython.display import Math, Latex
from IPython.core.display import Image
import seaborn as sns
sns.set(color_codes=True)
sns.set(rc={'figure.figsize':(8,6)})
from scipy.stats import uniform
number=10000
start=20
width=25
uniform_data=uniform.rvs(size=number,loc=start,scale=width)
axis = sns.distplot(uniform_data, bins=10, kde=True, color= 'skyblue', hist_kws={"linewidth":15})
axis.set(xlabel='Uniform Distribution', ylabel='Frequency')
```

**Distribusi Normal**


```python
from scipy.stats import norm
normal_data=norm.rvs(size=90000, loc=20, scale=30)
axis=sns.distplot(normal_data, bins=100, kde=True, color='skyblue', hist_kws={"linewidth":15,
                  'alpha':0.568})
axis.set(xlabel='Distribution Normal', ylabel='Frekuensi')
```

**DISTRIBUSI EKSPONENSIAL**
