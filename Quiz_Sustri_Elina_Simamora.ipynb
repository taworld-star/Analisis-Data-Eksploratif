```python
#7. Menghitung rata-rata data frame pertama kolom pertama
X = data['X'].mean()
print("Rata-rata kolom X=", X)
```

```python
#11. Grafik box plot
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.boxplot(data['X'],showmeans=True, whis=99)
plt.title('Box Plot of X')
plt.ylabel('Value')
plt.show()
```
