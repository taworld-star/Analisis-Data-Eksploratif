Indented Block

```python
#Let's say we have data on the heights of a group of students as for
```

```python
height=[120,122,125,127,121,123,137,131,161,145,141,132]
height
```

```python
bins =[118, 125, 135, 160, 200]
category = pd.cut(height, bins)
category
```

```python
category2=pd.cut(height, [118, 126, 136, 161, 200], right=False)
category2
```

Arti kurung yang berbeda, angka dikurung awal nilai tidak dimasukkan berbeda dengan angka setelah kurung siku , nilai dimasukkan.

Arti Kurung dalam Interval

(118, 125] :

( berarti nilai 118 tidak termasuk dalam interval.
] berarti nilai 125 termasuk dalam interval.
Jadi, interval ini mencakup semua nilai dari 118 (eksklusif) hingga 125 (inklusif).
Secara umum:

( berarti eksklusif (nilai tidak termasuk).
] berarti inklusif (nilai termasuk).

```python
pd.value_counts(category)
```

```python
#MENAMBAH LABEL
bin_names=['Short Height', 'Average height', 'Good Height', 'Taller']
pd.cut(height, bins, labels=bin_names)
```
